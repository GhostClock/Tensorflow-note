```python
import tensorflow as tf
```

```python
# 创建一个常量op
m1 = tf.constant([[3, 3]]) # 创建一行两列
# 创建一个常量op
m2 = tf.constant([[2], [3]]) # 创建两行一列

# 矩阵相乘的结果op, 把m1和m2传入
product = tf.matmul(m1, m2) # --> 3 * 2 + 3 * 3 = 15

print(product) ## 输出的是一个Tensor
```

```python
# 定义一个会话，启动默认的图
session = tf.Session()
# 调用session的run方法来执行矩阵乘法op
# run(product)触法图中3个op
result = session.run(product) # 这才是执行
print(result)

session.close() # 关闭会话
```

```python
# 一般这样定义一个会话 不需要手动关闭会话
with tf.Session() as session:
    result = session.run(product) # 这才是执行
    print(result)
```
